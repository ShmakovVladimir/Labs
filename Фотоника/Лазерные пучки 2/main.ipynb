```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
from scipy import optimize
from scipy import stats
plt.style.use(r"/home/vladimir/BOTAY!/Labs/jupyter_template/style.mplstyle")
```

### Красный лазер

```python
data = pd.read_csv("red.csv")
w = (data['x_left[mm/100]'] - data['x_right[mm/100]']) / 100


def parabola(x, a, b, c):
    return a * np.power(x + c, 2) + b

params, _ = optimize.curve_fit(parabola, data['Z[cm]'], w)


x_fit = np.linspace(0, 80, 1000)
y_fit = parabola(x_fit, *params)

plt.plot(x_fit, y_fit)
plt.scatter(data['Z[cm]'], w)
plt.xlabel("Расстояне [$см$]")
plt.ylabel("Шарина пучка [$мм$]")
print(f"Расстояние до точки перетяжки: {-params[2]:.2f}")
plt.savefig("red.png", bbox_inches = 'tight')
```
