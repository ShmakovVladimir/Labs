```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = [ 14, 16, 17, 18, 28, 32, 40]
```

```python
pressureExp = np.array([ 5, 4, 9, 16, 32, 8, 0])
pressureTable = np.array([0.5, 0.4, 0, 1.5, 4, 0.8, 3.5 ])
```

```python
pressureExpNorm = pressureExp/sum(pressureExp)
pressureTableNorm = pressureTable/sum(pressureTable)
```

```python
fig, ax = plt.subplots(1,2,figsize = (16,9), dpi = 400)

ax[0].set_title(",    ")
ax[0].pie(pressureExpNorm,labels = labels);

ax[1].set_title(" ")
ax[1].pie(pressureTableNorm,labels = labels);
```
